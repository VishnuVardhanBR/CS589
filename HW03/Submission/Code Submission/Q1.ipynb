# Question 1: Learning Probability Mass Functions with Symbolic Differentiation

```python
import sympy as sp
sp.init_printing(use_latex='mathjax')
```

### a-b. Define the NALL function

```python
pi, N0, N1, Ntr = sp.symbols('pi N_0 N_1 N_{tr}')
nall = -1/Ntr * (N1*sp.log(pi) + N0*sp.log(1-pi))
nall
```

### c. Derivative of the NALL Function

```python
derivative_nall = sp.diff(nall, pi)
simplified_derivative = sp.simplify(derivative_nall)
simplified_derivative
```

### d. Solving the Stationary Equation

```python
solution = sp.solve(simplified_derivative, pi)
solution[0]
```
