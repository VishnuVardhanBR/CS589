### Question 1: Bernoulli Mixture

```python
import numpy as np
```

### Load model and data

```python
data = np.load("bernoulli.npz")
phi  = data["phi"]
pi   = data["pi"]
x    = data["x"]
K,D  = phi.shape

print(K)
print(D)
```

### Question 1 (b,c)

```python
def log_joint_probability(x, pi, phi):

    x = x.reshape(1, -1)

    log_prior = np.log(pi).flatten()

    phi_clipped = np.clip(phi, 1e-10, 1 - 1e-10)

    log_likelihood = np.sum(x * np.log(phi_clipped) + (1 - x) * np.log(1 - phi_clipped), axis=1)  # Shape: (K,)

    log_joint = log_prior + log_likelihood

    return log_joint

log_joint = log_joint_probability(x, pi, phi)

print("Question 1(c): Values of log P(X=x, Z=k) for each k:")
print("-" * 50)
for k in range(K):
    print(f"log P(X=x, Z={k}) = {log_joint[k]:.6f}")
```

### Question 1 (d,e)

```python

def log_conditional_probability(x, pi, phi):

    log_joint = log_joint_probability(x, pi, phi)
    max_log_joint = np.max(log_joint)
    log_marginal = max_log_joint + np.log(np.sum(np.exp(log_joint - max_log_joint)))

    log_posterior = log_joint - log_marginal

    return log_posterior

log_posterior = log_conditional_probability(x, pi, phi)
posterior = np.exp(log_posterior)

print("Question 1(f): Values of log P(Z=k | X=x) and P(Z=k | X=x) for each k:")
print("-" * 60)
for k in range(K):
    print(f"log P(Z={k} | X=x) = {log_posterior[k]:.6f}  |  P(Z={k} | X=x) = {posterior[k]:.6f}")

print("-" * 60)
print(f"Sum of P(Z=k | X=x) = {np.sum(posterior):.6f}")
```
